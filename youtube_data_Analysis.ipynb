Data Cleaning  (Replace '--'  to NaN)

```python
df.head(20)
```

```python
df = df.replace('--', np.nan, regex=True)
```

```python
df.head(20)
```

Check Null Values In The Dataset

```python
df.isnull().sum()
```

```python
df.dropna(axis=0, inplace=True)
```

```python
df.isnull().sum()
```

Data Cleaning [ Rank Column ]

```python
df.head()
```

```python
df.tail()
```

```python
df['Rank'] = df['Rank'].str[:-2]
```

```python
df.head()
```

```python
df.tail()
```

```python
df['Rank'] = df['Rank'].str.replace(',','').astype('int')
```

```python
df.dtypes
```

Data Cleaning [ Video Uploads & Subscribers ]

```python
df['Video Uploads'] = df['Video Uploads'].astype('int')
df['Subscribers'] = df['Subscribers'].astype('int')
```

```python
df.dtypes
```

Data Cleaning [ Grade Column ]

```python
df.head()
```

```python
df['Grade'].unique()
```

```python
df['Grade'] = df['Grade'].map({'A++ ':5, 'A+ ':4, 'A ':3, 'A- ':2, 'B+ ':1})
```

```python
df.dtypes
```

Find Average Views For Each Channel

```python
df.head()
```

```python
df.columns
```

```python
df['Avg views']=df['Video views'] / df['Video Uploads']
```

```python
df.head()
```

Find Out Top Five Channels With Maximum Number of Video Uploads

```python
df.sort_values(by='Video Uploads', ascending=False).head()
```

 Find Correlation Matrix

```python
df.corr()
```

Which Grade Has A Maximum Number of Video Uploads?

```python
plt.figure(figsize=(15,10));
sns.barplot(data=df, x='Grade', y='Video Uploads');
plt.title('Grade vs Video Uploads');
plt.xlabel('Grade', fontsize=13);
plt.ylabel('Video Uploads', fontsize=13);
```

Which Grade Has The Highest Average Views?

```python
df.head()
```

```python
plt.figure(figsize=(15,10));
sns.barplot(data=df, x='Grade', y='Avg views');
plt.title('Grade vs Avg views');
plt.xlabel('Grade', fontsize=13);
plt.ylabel('Avg views', fontsize=13);
```

 Which Grade Has The Highest Number of Subscribers?

```python
plt.figure(figsize=(15,10));
sns.barplot(data=df, x='Grade', y='Subscribers');
plt.title('Grade vs Subscribers');
plt.xlabel('Grade', fontsize=13);
plt.ylabel('Subscribers', fontsize=13);
```

Which Grade Has The Highest Video Views? 

```python
plt.figure(figsize=(15,10));
sns.barplot(data=df, x='Grade', y='Video views');
plt.title('Grade vs Video views');
plt.xlabel('Grade', fontsize=13);
plt.ylabel('Video views', fontsize=13);
```
